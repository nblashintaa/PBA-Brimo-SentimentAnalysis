```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

corpus = [
    "Bank Mandiri menyalurkan pembiayaan berkelanjutan melalui penerapan ESG untuk mendukung ekonomi hijau.",
    "Bank Mandiri memperkuat penyaluran kredit berkelanjutan dalam segmen wholesale dengan mengintegrasikan ESG.",
    "Bank Mandiri mendukung transisi energi dan ekonomi rendah karbon melalui pembiayaan berkelanjutan dan perdagangan karbon."
]

vectorizer = CountVectorizer(
    vocabulary=['bank', 'mandiri', 'kredit', 'pembiayaan', 'berkelanjutan', 'esg', 'energi', 'karbon']
)

X = vectorizer.fit_transform([text.lower() for text in corpus])

cos_sim = cosine_similarity(X[0:1], X)

df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df['Cosine Sim to Doc 0'] = cos_sim.flatten()

print(df)
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
