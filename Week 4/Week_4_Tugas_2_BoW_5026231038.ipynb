```python
import nltk
# Download dataset yang dibutuhkan NLTK

# Stopwords = kata umum yang tidak penting (the, is, and, dll)
nltk.download('stopwords')

# Tokenizer kalimat
nltk.download('punkt_tab')

# Kamus lemmatization (mengubah kata ke bentuk dasar)
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer # Mengubah kata ke bentuk dasar
from nltk.corpus import stopwords # Daftar kata umum (stopwords)
import re
paragraph = "Wakil Direktur Bank Mandiri Alexandra Askandar menyatakan hingga kuartal ketiga 2023, sebanyak Rp 122 triliun digunakan untuk pembiayaan Green Portfolio dan senilai Rp131 triliun untuk pembiayaan Social Portfolio. Sehingga Sustainable Loan Portfolio 2023 Bank Mandiri berkontribusi sebesar 24,9% dari total pinjaman perseroan."

# Tokenisasi
sentences = nltk.sent_tokenize(paragraph)
lemmatizer = WordNetLemmatizer()
corpus = []

# Preprocessing tiap kalimat
for i in range(len(sentences)):
    sent = re.sub('[^a-zA-Z]', ' ', sentences[i])
    sent = sent.lower()
    sent= sent.split()
    sent = [lemmatizer.lemmatize(word) for word in sent if not word in set(stopwords.words('english'))]
    sent = ' '.join(sent)
    corpus.append(sent)
print(corpus)
```

List all unique words

```python
print("Corpus:", corpus)
```

```python
# Finding unique words
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print("Unique words:", unique_words)
```

Membuat sebuah kamus (dictionary) yang berisi : kata (word) dipasangkan dengan angka

```python
from collections import Counter

all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

# Menghitung frekuensi kemunculan setiap kata
word_frequencies = Counter(all_words)

# Menampilkan jumlah kemunculan tiap kata
print("Word Frequencies:", word_frequencies)

# Mengurutkan kata berdasarkan frekuensi (terbanyak → terendah)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Membuat mapping kata ke angka (vocabulary indexing)
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print("Word to Number Mapping (Sorted by Frequency):", word_to_number)
```
