```python
import nltk
from nltk.tokenize import word_tokenize
import math

# Download the 'punkt' data if needed
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

sentences = [
    "Manchester City makes history by winning Club World Cup.",
    "Manchester City capped off its incredible year with yet another trophy.",
    "Pep Guardiola side now boasts five trophies this calendar year.",
    "Manchester City outclassed its Brazilian opponents in the final.",
    "The team showed consistency and strong performance.",
    "Manchester City took the lead just 40 seconds into the game.",
    "City continued to look dangerous before the break.",
    "Foden got on the scoresheet in the 72nd minute.",
    "Alvarez completed the rout with a brilliant performance.",
    "City defence was largely untested during the game."
]

all_cleaned_sentences = []
for sentence in sentences:
    cleaned_sentence = sentence.lower()
    all_cleaned_sentences.append(cleaned_sentence)

def calculate_tf(term, document):
    words = word_tokenize(document.lower())
    return words.count(term) / len(words) if len(words) > 0 else 0

def calculate_idf(term, all_documents):
    num_documents_with_term = sum(
        1 for doc in all_documents if term.lower() in word_tokenize(doc.lower())
    )
    if num_documents_with_term > 0:
        return math.log(len(all_documents) / num_documents_with_term)
    else:
        return 0

target_word = "manchester"

# Calculate TF-IDF
tfidf_scores = {}
for i, sentence in enumerate(sentences):
    cleaned_sentence = sentence.lower()
    tf = calculate_tf(target_word, cleaned_sentence)
    idf = calculate_idf(target_word, all_cleaned_sentences)
    tfidf = tf * idf
    tfidf_scores[f"Sentence {i+1}"] = tfidf

# Print
print(f"TF-IDF Score for '{target_word}' in each sentence:")
for sentence_num, score in tfidf_scores.items():
    print(f"{sentence_num}: {score:.4f}")

# Overall IDF
overall_idf = calculate_idf(target_word, all_cleaned_sentences)
print(f"\nOverall IDF for '{target_word}': {overall_idf:.4f}")
```
